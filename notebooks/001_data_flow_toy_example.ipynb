# Toy Data Flow

This notebook mirrors the runtime audit trail without needing Docker, Alpaca, Postgres, pandas, or sklearn.

The real system stores each step in Postgres:

```text
market snapshot -> features -> inference -> decision -> optional paper order -> outcome
```

````python
market_snapshot = {
    "symbol": "SPY",
    "timeframe": "1Min",
    "open": 100.0,
    "high": 101.0,
    "low": 99.5,
    "close": 100.5,
    "volume": 12345,
}

features = {
    "symbol": market_snapshot["symbol"],
    "close": market_snapshot["close"],
    "returns_1": 0.002,
    "returns_5": 0.006,
    "moving_average_distance_20": 0.004,
    "volatility_20": 0.001,
}

inference = {
    "predicted_action": "buy",
    "confidence": 0.61,
    "model_name": "toy-baseline",
    "model_version": "not-trained",
}

decision = {
    "symbol": market_snapshot["symbol"],
    "proposed_action": inference["predicted_action"],
    "proposed_qty": 1,
    "policy_status": "approved",
    "final_action": "buy",
}

audit_row = {
    "market_snapshot": market_snapshot,
    "features": features,
    "inference": inference,
    "decision": decision,
}

audit_row
````

The point of the project is not the toy `buy` output above. The point is that every row has enough context to later ask: did this decision help?
